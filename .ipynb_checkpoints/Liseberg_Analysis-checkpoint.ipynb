```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import scipy.fftpack
```

Fetch the data prepared by the prepare data script

```python
df = pd.read_csv('./dataframes/data.csv', index_col=0)
```

Create a new dataframe for the rollecoasters and print it's head and plot it.

```python
extreme_rides = df[['Balder', 'Valkyria', 'Lisebergbanan','Helix',
                    'AtmosFear', 'FlumeRide', 'Hanghai', 'Kaninlandsbanan','Loke']].copy()
plt.figure(figsize=(16, 6))
extreme_rides.plot()
```

Check the sum of the columns in the dataframe and plot it

```python
sum = extreme_rides.sum(1)
plt.figure(figsize=(16, 6))
sum.plot()
```

Ok, so this looks interesting. The sum of the queue time is fairly constant between the "high load times"

```python
plt.figure(figsize=(16, 6))
fig = extreme_rides.hist()
```

Take a look at the correlation of the "extreme rides"

```python
corr = extreme_rides.corr()
plt.figure(figsize=(16, 6))
fig = plt.matshow(corr)
print(corr)
```
